4. Write a Python program to read last n lines of a file.

```python
n = 1

with open('Text Files/sample.txt', 'r') as f:
    lines = f.readlines()

for line in lines[-n:]: # use slicing
    print(line, end='')
```

5. Write a Python program to read a file line by line and store it into a list.

```python
L =[]

with open('Text Files/sample.txt', 'r') as f:
    while True:
        data = f.readline()
        if data == "":
            break
        else:
            L.append(data)

print(L)
```

6. Write a Python program to read a file line by line store it into a variable.

```python
a = ""
with open('Text Files/sample.txt','r') as f:
    while True:
        data = f.readline()

        if data == "":
            break
        else:
            a += data
print(a)
```

7. Write a python program to find the longest words.

```python
with open("Text Files/sample.txt",'r') as f:
    for line in f:
        parts = line.strip().split()

        a = parts[0]
        b = parts[1]

        temp = a
        a = b
        b = temp
```

8. Write a Python program to count the number of lines in a text file.

```python
with open('Text Files/sample.txt','r') as f:
    count_line = 0
    for line in f:
        count_line += 1
    

print(count_line)
```

9. Write a Python program to count the frequency of words in a file.

```python
d = {}

with open('Text Files/sample.txt','r') as f:
    for line in f:
        words = line.lower().split()
        for word in words:
            if word in d:
                d[word] += 1
            else:
                d[word] = 1

print(d)
```

15. Write a Python program that takes a text file as input and returns the number of words of a given text file.

```python
with open('Text Files/sample.txt','r') as f:
    count = 0
    for line in f:
        words = line.split()
        for word in words:
            count += 1
print(count)
```
